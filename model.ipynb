```python

import pandas as pd
import pickle
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
#from lightgbm import LGBMClassifier
#from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score, f1_score,recall_score,roc_auc_score
```

```python
#Import the dataset
df = pd.read_csv("cleaned_heart_data.csv")
df.head()
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

Train Test Split

```python
X = df.drop(['Heart Attack', 'Blood Pressure(mmHg)'], axis=1)
y = df['Heart Attack']
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Model Building

### LogisticRegression

```python
lr = LogisticRegression(penalty='l2')
lr.fit(X_train, y_train)
print(lr.coef_)
print(lr.classes_)
print(lr.intercept_)
```

```python
# Training Metrics
y_train_pred = lr.predict(X_train)
print(roc_auc_score(y_train, y_train_pred))
print(recall_score(y_train, y_train_pred))
confusion_matrix(y_train, y_train_pred)
print(classification_report(y_train, y_train_pred))
```

```python
# Test Metrics
y_test_pred = lr.predict(X_test)
print(roc_auc_score(y_test, y_test_pred))
print(recall_score(y_test, y_test_pred))
print(accuracy_score(y_test, y_test_pred))
print(confusion_matrix(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

```python
model_pkl_file = "logistic_regression_model.pkl"
with open(model_pkl_file,'wb') as file:
    pickle.dump(lr, file)
```

```python
import joblib

filename = 'joblib_logistic_reg.sav'
joblib.dump(lr,filename)
```

## Decision Tree 

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
```

```python
# Training Metrics
y_train_pred = dt.predict(X_train)
print(roc_auc_score(y_train, y_train_pred))
print(recall_score(y_train, y_train_pred))
confusion_matrix(y_train, y_train_pred)
print(classification_report(y_train, y_train_pred))
```

```python
# Test Metric
y_test_pred = dt.predict(X_test)
print(roc_auc_score(y_train, y_train_pred))
print(recall_score(y_train, y_train_pred))
print(confusion_matrix(y_train, y_train_pred))
print(classification_report(y_train, y_train_pred))
```
